```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()

# Nodes with heuristic values
nodes = {"X": 5, "Y": 6, "W": 4, "V": 8, "U": 7, "T": 5, "Z": 9}
edges = [("X", "Y", 2), ("X", "W", 3), ("Y", "V", 2), ("Y", "U", 3),
         ("W", "T", 4), ("V", "Z", 3), ("U", "Z", 6), ("T", "Z", 2)]

G.add_nodes_from(nodes.keys())
G.add_weighted_edges_from(edges)

pos = {"X": (0, 3), "Y": (2, 4), "W": (2, 2), "V": (4, 5),
       "U": (4, 3), "T": (4, 1), "Z": (6, 4)}

plt.figure(figsize=(8, 5))
nx.draw(G, pos, with_labels=True, node_size=2000, node_color="lightblue", font_size=10, edge_color="gray")
edge_labels = {(src, dst): f"{cost}" for src, dst, cost in edges}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=9, font_color="red")
node_labels = {node: f"{node}\n(h={heuristic})" for node, heuristic in nodes.items()}
nx.draw_networkx_labels(G, pos, labels=node_labels, font_size=9, font_color="darkblue")

plt.title("Hill Climbing Algorithm Graph")
plt.show()
```

```python
G = nx.DiGraph()

nodes = {"A": 6, "B": 4, "C": 5, "D": 2, "E": 3, "F": 2, "G": 0}
edges = [("A", "B", 1), ("A", "C", 4), ("B", "D", 3), ("B", "E", 1),
         ("C", "F", 5), ("D", "G", 2), ("E", "G", 5), ("F", "G", 2)]

G.add_nodes_from(nodes.keys())
G.add_weighted_edges_from(edges)

pos = {"A": (0, 3), "B": (2, 4), "C": (2, 2), "D": (4, 5),
       "E": (4, 3), "F": (4, 1), "G": (6, 3)}

plt.figure(figsize=(8, 5))
nx.draw(G, pos, with_labels=True, node_size=2000, node_color="lightgreen", font_size=10, edge_color="gray")
edge_labels = {(src, dst): f"{cost}" for src, dst, cost in edges}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=9, font_color="red")
node_labels = {node: f"{node}\n(h={heuristic})" for node, heuristic in nodes.items()}
nx.draw_networkx_labels(G, pos, labels=node_labels, font_size=9, font_color="darkblue")

plt.title("A* Algorithm Graph")
plt.show()
```
